```python
#import necessary packages
import geopandas as gpd
import pandas as pd
import numpy as np
import folium as flm
from folium import plugins
from folium.plugins import HeatMap
```

```python
#read the Glascontainer file
df=pd.read_json('Glascontainer.json')
df_bins=pd.DataFrame(data={'zip': df['zip'], 'address':df['address'], 'lat':df['lat'], 'lng': df['lng']})
df_bins['zip'].replace('', np.nan, inplace=True)
df_bins.dropna(subset=['zip'], inplace=True)
#reset index
df_bins.reset_index(drop=True, inplace=True)
df_bins['address']=df['address'].str.replace('`','')
df_bins
```

```python
#read the Shapefile of berlin
berlin='Berlin_Bezirke.shp'
berlin_map=gpd.read_file(berlin)
```

```python
from folium.plugins import HeatMap

for_map = df_bins
hmap = flm.Map(location=[52.5, 13.5], zoom_start=10, )

hm_wide = HeatMap( list(zip(for_map['lat'], for_map['lng'])),
                   min_opacity=1,max_opacity=10,
                   radius=13, blur=10, 
                   max_zoom=10)
marker_cluster=plugins.MarkerCluster().add_to(hmap)
for i in range(len(for_map)):
    flm.CircleMarker(location=[for_map['lat'].iloc[i],for_map['lng'].iloc[i]], radius=10, tooltip=for_map['address'].iloc[i], color='#FFBA00', fill_color='FFBA00', fill=True).add_to(marker_cluster)
                 
flm.GeoJson(berlin_map).add_to(hmap)
hmap.add_child(hm_wide)
```
